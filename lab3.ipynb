```python
import tensorflow as tf
import numpy as np
```

```python
from tensorflow.keras.layers import (
    Input,
    Embedding,
    Flatten,
    Dot,
    Subtract
)
from tensorflow.keras.models import Model
```

```python
num_users = 5
num_items = 10
embedding_dim = 8
```

```python
user_input = Input(shape=(1,))
positive_item_input = Input(shape=(1,))
negative_item_input = Input(shape=(1,))
```

```python
user_embedding = Embedding(
    input_dim=num_users,
    output_dim=embedding_dim
)(user_input)

positive_embedding = Embedding(
    input_dim=num_items,
    output_dim=embedding_dim
)(positive_item_input)

negative_embedding = Embedding(
    input_dim=num_items,
    output_dim=embedding_dim
)(negative_item_input)
```

```python
user_vec = Flatten()(user_embedding)
pos_vec = Flatten()(positive_embedding)
neg_vec = Flatten()(negative_embedding)
```

```python
positive_score = Dot(axes=1)([user_vec, pos_vec])
negative_score = Dot(axes=1)([user_vec, neg_vec])
```

```python
output = Subtract()([positive_score, negative_score])

# Build Model
model = Model(
    inputs=[
        user_input,
        positive_item_input,
        negative_item_input
    ],
    outputs=output
)
```

```python
model.compile(
    optimizer='adam',
    loss='mse'
)
```

```python
users = np.array([0, 1, 2, 3])
positive_items = np.array([1, 2, 3, 4])
negative_items = np.array([5, 6, 7, 8])

target = np.ones((4, 1))
```

```python
model.fit(
    [users, positive_items, negative_items],
    target,
    epochs=20
)

print("Model trained successfully")
```
